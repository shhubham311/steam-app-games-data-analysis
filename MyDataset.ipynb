# Distributions

```python
from scipy.stats import norm, skew, kurtosis

# Assuming you have loaded the data into a DataFrame named 'df'
data = df['price']

# Create a histogram
plt.hist(data, bins=50, alpha=0.6, color='g', density=True)

# Fit a normal distribution to the data
mu, std = norm.fit(data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# Plot the normal distribution curve
plt.plot(x, p, 'k', linewidth=2)
title = "Fit Results: Mean = %.2f, Std = %.2f" % (mu, std)
plt.title(title)

# Check for normality
skewness = skew(data)
kurt = kurtosis(data)
print('Skewness: ', skewness)
print('Kurtosis: ', kurt)

# Perform normal distribution transformation
transformed_data = norm.ppf(norm.cdf(data))

# Print the transformed data
print('Transformed data: ', transformed_data)


# Display the histogram
plt.show() 
```

By analyzing the skewness and kurtosis values, we can determine the type of distribution our dataset follows. Then, by using the inverse cumulative distribution function (ICDF) provided by the norm.ppf method, we can transform our data to follow a normal distribution.

# Hypothesis Testing 


**Hypothesis: The number of positive ratings (upvotes) a game receives is higher for more recent games compared to older games.**

```python
# Visualizing the distribution of upvotes for recent games

recent_games = df[df['release_date'] > '2017']
plt.hist(recent_games['positive_ratings'], bins=50, alpha=0.5, label='Recent Games')

# Visualizing the distribution of upvotes for older games

older_games = df[df['release_date'] <= '2017']
plt.hist(older_games['positive_ratings'], bins=50, alpha=0.5, label='Older Games')

plt.title('Distribution of Upvotes')
plt.xlabel('Upvotes')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```

**Performing a Two-Sample T-Test to check if the number of positive ratings is higher for recent games.**

```python
from scipy.stats import ttest_ind

games_before_2017 = df[df['release_date']< '2017']['positive_ratings']
games_after_2017 = df[df['release_date'] >= '2017']['positive_ratings']

t_statistic, p_value = ttest_ind(games_before_2017, games_after_2017)

print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis. The number of positive ratings is higher for recent games.")
else:
    print("Cannot reject the null hypothesis. There is no significant difference in the number of positive ratings between recent and non-recent games.")
```

**Performing a chi squared test to check if the distribution of games based on their average playtime is the same as the distribution of games based on their median playtime.**

```python
from scipy.stats import chi2_contingency

df.drop_duplicates(subset='median_playtime', inplace=True)
df.drop_duplicates(subset='average_playtime', inplace=True)

df['average_playtime_decile'] = pd.qcut(df['average_playtime'], 10, labels=False)
df['median_playtime_decile'] = pd.qcut(df['median_playtime'], 10, labels=False)

average_playtime_decile_games = df.groupby('average_playtime_decile')['appid'].count()
median_playtime_decile_games = df.groupby('median_playtime_decile')['appid'].count()

chi2, p_value, _, _ = chi2_contingency(average_playtime_decile_games, median_playtime_decile_games)

print(f"Chi-Squared Statistic: {chi2}")
print(f"P-Value: {p_value}")


# Defining the significance level
alpha = 0.05

if p_value < alpha:
    print('Reject the null hypothesis: The distribution of games based on their average playtime is the same as the distribution of games based on their median playtime.')
else:
    print('Fail to reject the null hypothesis: The distribution of games based on their average playtime is not the same as the distribution of games based on their median playtime.')
```
