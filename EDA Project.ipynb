# Distributions

```python

from scipy.stats import norm

# Assuming you have loaded the data into a DataFrame named 'df'
data = df['price']

# Create a histogram
plt.hist(data, bins=50, alpha=0.6, color='g', density=True)

# Fit a normal distribution to the data
mu, std = norm.fit(data)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# Plot the normal distribution curve
plt.plot(x, p, 'k', linewidth=2)
title = "Fit Results: Mean = %.2f, Std = %.2f" % (mu, std)
plt.title(title)

# Display the histogram
plt.show()
```

**Checking the type of distribution and convert it to normal distribution:**

```python
from scipy.stats import norm, kurtosis, skew

# Check for normality
skewness = skew(data)
kurt = kurtosis(data)
print('Skewness: ', skewness)
print('Kurtosis: ', kurt)

# Perform normal distribution transformation
transformed_data = norm.ppf(norm.cdf(data))
```

By analyzing the skewness and kurtosis values, you can determine the type of distribution your dataset follows. Then, by using the inverse cumulative distribution function (ICDF) provided by the norm.ppf method, you can transform your data to follow a normal distribution.

# Hypothesis Testing 


**Hypothesis: The number of positive ratings (upvotes) a game receives is higher for more recent games compared to older games.**

```python
# Visualize the distribution of upvotes for recent games
recent_games = df[df['release_date'] > '2017']
plt.hist(recent_games['positive_ratings'], bins=50, alpha=0.5, label='Recent Games')

# Visualize the distribution of upvotes for older games
older_games = df[df['release_date'] <= '2017']
plt.hist(older_games['positive_ratings'], bins=50, alpha=0.5, label='Older Games')

plt.title('Distribution of Upvotes')
plt.xlabel('Upvotes')
plt.ylabel('Frequency')
plt.legend(loc='upper right')
plt.show()
```

**Performing a Two-Sample T-Test to check if the average number of positive reviews is higher for recent games.**

```python
# Check if the distributions are approximately normal
_, recent_games_p = stats.normaltest(recent_games['positive_ratings'])
_, older_games_p = stats.normaltest(older_games['positive_ratings'])

# Perform the T-Test if the distributions are normal
if recent_games_p > 0.05 and older_games_p > 0.05:
    t_stat, p_value = stats.ttest_ind(recent_games['positive_ratings'], older_games['positive_ratings'])
    print(f"T-statistic: {t_stat}, P-value: {p_value}")
else:
    print("The distributions are not approximately normal. Use a non-parametric test instead.")
```

```python

from scipy.stats import ttest_ind

games_before_2017 = df[df['release_date']< '2017']['positive_ratings']
games_after_2017 = df[df['release_date'] > '2017']['positive_ratings']

t_statistic, p_value = ttest_ind(games_before_2017, games_after_2017)

print(f"T-Statistic: {t_statistic}")
print(f"P-Value: {p_value}")

if p_value < 0.05:
    print("Reject the null hypothesis. The average number of positive reviews is higher for recent games.")
else:
    print("Cannot reject the null hypothesis. There is no significant difference in the average number of positive reviews between recent and non-recent games.")
```

# Findings and Insights

* Games with Longer Release Dates and Lower Positive Review Counts: We found that games with longer release dates and lower positive review counts were more likely to be priced higher. This could indicate that Steam prioritizes newer releases or more established titles, potentially resulting in higher prices for older games.

* Relatively Low Number of Low-Price Games with Many Reviews: Our analysis also revealed a relatively low number of low-price games with a high number of reviews. This could suggest that Steam may not be actively promoting low-price games with many reviews, potentially due to concerns about diluting the positive reviews for high-priced games.

In conclusion, our analysis of the Steam store games dataset has provided valuable insights into the relationships between different variables.
